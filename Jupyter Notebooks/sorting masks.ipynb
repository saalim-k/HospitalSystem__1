```python
import pickle
import cv2
import matplotlib.pyplot as plt
import numpy as np

X_train= pickle.load(open("Datasets/binaryFormNorm/X_train.pickle","rb"))
Y_train = pickle.load(open("Datasets/binaryFormNorm/y_train.pickle","rb"))

```

```python
fig=plt.figure(figsize=(20,50))
rows = 1
columns = 5
i=0
for image,label in zip(X_train[0:5],Y_train[0:5]):
    i+=1
    fig.add_subplot(rows, columns, i)
    plt.imshow(image )
    plt.title("Orig Img"+f'{label}' )
    fig.tight_layout(pad=1)
```
